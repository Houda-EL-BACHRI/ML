```python
#importation des bibliothéques
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.model_selection import cross_val_score
```

```python
data=pd.read_csv("http://math.univ-lyon1.fr/~gciuperca/data_sommeil.csv",sep=';')
print("Notre data est:")
data
```

```python
# Discretisez les valeurs des colones Age, Poids et ecran.
data["AGEQ"]=pd.qcut(data.AGE,3,labels=["AGE1","AGE2","AGE3"])
data["POIDSQ"]=pd.qcut(data.POIDS,3,labels=["POIDS1","POIDS2","POIDS3"])
data["ECRANQ"]=pd.qcut(data.ECRAN,4,labels=["ECRAN1","ECRAN2","ECRAN3","ECRAN4"])
```

```python
caractéres1=["AGE","TAILLE","POIDS","ALCOOL","ECRAN","TABAC","INSOMNIE","SEXE"]
for i in caractéres1:
    data[i]=pd.Categorical(data[i],ordered=True)
print(data.dtypes)
```

7)Modéles du ML :

  
*   Entrainer votre modèle

*   calculer les mesures de performance incluant :
    * -l'Accuracy
    * -la matrice de confusion


  

```python
X=["AGE", "POIDS", "TAILLE", "ALCOOL", "TABAC", "ECRAN"]
X = data[X]
Y=["INSOMNIE"]
Y = data[Y]

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,train_size=0.8)
```

naive bayésienne

```python
#la méthode naive bayésienne (GuassianNB)
model_Gaussian = GaussianNB()
# c)Entrainement du modèle
model_Gaussian.fit(x_train,y_train.values.ravel())
y_pred_GM = model_Gaussian.predict(x_test)
accuracy_GaussianNB = metrics.accuracy_score(y_test, y_pred_GM)*100
print("La précision du model est: ",accuracy_GaussianNB,'%')
print("La matrice de confusion:\n ",confusion_matrix(y_test,y_pred_GM))
```

Support Vector Machine (SVC).

```python
#modéle svm 
model = svm.SVC()
model.fit(x_train, y_train.values.ravel())
y_pred_svm = model.predict(x_test)

accuracy_SVC=metrics.accuracy_score(y_test, y_pred_svm)*100
print("la Précision du model est:",accuracy_SVC,'%')
print("La matrice de confusion:\n ",confusion_matrix(y_test,y_pred_svm))
```

le meilleur algorithme parmi : Regression Logistique, Naive Byésien, Support Vector MAchine (SVC).

```python
#modele LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train.values.ravel())
y_pred_LR = model.predict(x_test)
accuracy_LogisticRegression=metrics.accuracy_score(y_test, y_pred_LR)*100
print("la Précision du model est:",accuracy_LogisticRegression,'%')
print("La matrice de confusion:\n ",confusion_matrix(y_test,y_pred_LR))
```
